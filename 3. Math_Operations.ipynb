#Importing Libraries:

```python
import torch
import numpy as np
```

#Tensor Math Operations:

```python
t1 = torch.tensor([1,2,3,4])
t2 = torch.tensor([5,6,7,8])
#(Tensors must be of same size for math operations)
```

```python
#Addition:
print(t1 + t2)

#Alternative method:
print(torch.add(t1,t2))

t1.add(t2)
```

```python
#Subtraction:
print(t2-t1)

#Alternative:
print(torch.sub(t2,t1))

t2.sub(t1)
```

```python
#Multiplication:
print(t1*t2)

#Alternative:
print(torch.mul(t1,t2))

t1.mul(t2)
```

```python
#Division:
print(t2/t1)

#Alternative:
print(torch.div(t2,t1))

t2.div(t1)
```

```python
#Modulus/Remainder:
print(t2%t1)

#Alternative:
print(torch.remainder(t2,t1))

t2.remainder(t1)
```

```python
#Exponents/Power:
print(t1**t2)

#Alternative:
print(torch.pow(t1,t2))

t1.pow(t2)
```

```python
#Reassignment:
print(t1 + t2)
print("Before Reassignment:", t1)

#_() for reassignment
t1.add_(t2)
print("After Reassignment:", t1)

#OR:
#t1 = t1 + t2

#Similarly: .sub_, .mul_, .div_
```

#Importing Libraries:

```python
import torch
```

#Backpropagation:

```python
x = torch.tensor(1.0)
x
```

```python
y = torch.tensor(2.0)
y
```

```python
w = torch.tensor(1.0, requires_grad=True)
w
```

```python
#forward pass and compute loss
y_hat = w * x
loss = (y_hat - y)**2
print(loss)
```

```python
#backward pass
loss.backward()
print(w.grad)
```

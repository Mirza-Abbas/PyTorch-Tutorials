#Importing Libraries:

```python
import torch
import numpy as np
```

#Tensor Operations:

```python
t = torch.tensor([1,2,3])

if torch.cuda.is_available(): #Checks if cuda is available
    device = torch.device("cuda") #Assigns cuda to device var
    x = torch.ones(5, device=device) #Creates a tensor on gpu
    t = t.to(device) #moves tensor to cuda
    x = x.to("cpu") #moves tensor to cpu
else:
    device = torch.device("cpu")  #Assigns cpu to device var

print(device)
print(t)
```

```python
#Note: Numpy can only handle CPU tensors, can't convert GPU tensors

a = np.random.rand(3)

b = torch.tensor(a)   #Converts numpy array to tensor

print(b, type(b))

c = b.numpy()     #Converts tensor to numpy array

print(c, type(c))
```

```python
n1 = np.ones(4)

print(n1)

t1 = torch.from_numpy(n1)

print(t1)

n1 += 1 #Updating the numpy array will also update the tensor created from it (both share same memory)

print(n1)
print(t1)
```

```python
myt = torch.arange(10)
#creates a 1D tensor of 10 integers 0-9

myt
```

```python
myt = myt.reshape(2, 5)
#Changes the shape of tensor to 2x5

myt
```

```python
myt2 = torch.arange(14)

print(myt2)

myt2 = myt2.reshape(2, -1)
#Reshape without specifying number of items i.e. use -1

print(myt2)

myt2 = myt2.reshape(-1, 2)

myt2
```

```python
myt3 = torch.arange(6)

myt3 = myt3.view(3,2)
#Works similar to reshape

myt3
```

```python
#updating the original tensor also updates the tensors created using the original tensor

t1 = torch.arange(12)
t2 = t1
print("Before Updating t1", t2)
t1[1] = 10
print("After Updating t2", t2)
```

```python
myt4 = torch.arange(10)
myt4 = myt4.reshape(2,5)

print(myt4)

myt5 = myt4[:,1]
#Creates a 1D tensor with only column 1 items
print(myt5)

myt5 = myt4[1,:]
#Creates a 1D tensor with only row 1 items
print(myt5)

myt5 = myt4[:,1:]
#Creates a tensor having same structure including all columns from column 1
myt5
```
